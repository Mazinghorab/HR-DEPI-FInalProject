```python
# 1. Load all tables
emp = pd.read_csv('Employee.csv')
per = pd.read_csv('PerformanceRating.csv')
edu_level = pd.read_csv('EducationLevel.csv')
rating_level = pd.read_csv('RatingLevel.csv')
sat_level = pd.read_csv('SatisfiedLevel.csv')
```

```python
print(f"Loaded Tables:")
print(f"  Employee          : {emp.shape}")
print(f"  PerformanceRating : {per.shape}")
print(f"  EducationLevel    : {edu_level.shape}")
print(f"  RatingLevel       : {rating_level.shape}")
print(f"  SatisfiedLevel    : {sat_level.shape}")
```

```python
# 2. Check duplicates & nulls
print(f"Null Values in Employee : {emp.isnull().sum().sum()}")
print(f"Duplicates in Employee : {emp.duplicated().sum()}\n")

print(f"Null Values in PerformanceRating : {per.isnull().sum().sum()}")
print(f"Duplicates in PerformanceRating : {per.duplicated().sum()}\n")

print(f"Null Values in EducationLevel : {edu_level.isnull().sum().sum()}")
print(f"Duplicates in EducationLevel : {edu_level.duplicated().sum()}\n")

print(f"Null Values in RatingLevel : {rating_level.isnull().sum().sum()}")
print(f"Duplicates in RatingLevel : {rating_level.duplicated().sum()}\n")

print(f"Null Values in SatisfiedLevel : {sat_level.isnull().sum().sum()}")
print(f"Duplicates in SatisfiedLevel : {sat_level.duplicated().sum()}\n")
```

```python
# 3. Parse dates
emp['HireDate'] = pd.to_datetime(emp['HireDate'])
per['ReviewDate'] = pd.to_datetime(per['ReviewDate'])
```

```python
# 4. Fix string whitespace in lookup tables
edu_level['EducationLevel']   = edu_level['EducationLevel'].str.strip()
sat_level['SatisfactionLevel']  = sat_level['SatisfactionLevel'].str.strip()
rating_level['RatingLevel']     = rating_level['RatingLevel'].str.strip()
```

```python
# 5. Build master dataframe
master = emp.merge(
    per, on = 'EmployeeID', how = 'left'
).merge(
    edu_level, left_on = 'Education', right_on = 'EducationLevelID', how = 'left'
).merge(
    rating_level.rename(columns = {'RatingID':'SelfRating','RatingLevel':'SelfRatingLabel'}),
    on = 'SelfRating', how = 'left'
).merge(
    rating_level.rename(columns = {'RatingID':'ManagerRating','RatingLevel':'ManagerRatingLabel'}),
    on = 'ManagerRating', how = 'left'
).merge(
    sat_level.rename(columns = {'SatisfactionID':'JobSatisfaction','SatisfactionLevel':'JobSatLabel'}),
    on = 'JobSatisfaction', how = 'left'
)
```

```python
master.head()
```

```python
# 6. Derived columns
master['AttritionBinary'] = (master['Attrition'] == 'Yes').astype(int)
master['SalaryBand'] = pd.cut(master['Salary'],
    bins = [0, 50000, 80000, 110000, master['Salary'].max() + 1],
    labels = ['Low', 'Medium', 'High', 'Very High'])
```
